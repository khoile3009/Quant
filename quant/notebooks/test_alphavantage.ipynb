```python
import pandas as pd

class AlphaVantage:
    ALPHA_VANTAGE_API_KEY = "5QQ3NMDSVUMS1EDX"
    BASE_API_PATH = "https://www.alphavantage.co/query?"
    
    def query(self,*args, **kwargs):
        # https://www.alphavantage.co/documentation/
        if(not kwargs.get("datatype")):
            kwargs["datatype"] = "csv"
        kwargs["apikey"] = self.BASE_API_PATH
        return pd.read_csv(self.get_query_path(kwargs))
    
    def get_query_path(self,kwargs):
        resulted_query_path = self.BASE_API_PATH
        for key, value in kwargs.items():
            resulted_query_path += "{0}={1}&".format(key,value)
        print(resulted_query_path)
        return resulted_query_path

AV = AlphaVantage()
AV.query(function="TIME_SERIES_DAILY", symbol="IBM",)    
```
